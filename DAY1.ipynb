```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=np.exp(x)
plt.plot(x,y,marker='o',label='exp function')
plt.loglog(x,x**2,marker='^',label='$x^2$ function')#loglog
plt.xlabel('x')
plt.ylabel(r'$y=e^x$')
plt.title("Exponential function")
plt.legend()
plt.grid()
plt.show()
```

```python
#can be used for curve fittings
t=np.random.randn(5000)
s=np.sin(np.pi*t)+0.1*np.random.randn(5000)
p=np.cos(np.pi*t)+0.5*np.random.randn(5000)
plt.scatter(t,s,s=1,color='r',marker='.')
plt.scatter(t,p,s=1,color='g',marker='.')
```

```python
x=np.linspace(0,2*np.pi,1000)
sinx=np.sin(x)
sin2x=np.sin(x)**2
cosx=np.cos(x)
cos2x=np.cos(x)**2
```

```python
fig,ax=plt.subplots(2,2,figsize=(16,9),dpi=100)
ax[0,0].plot(x,sinx)


ax[0,1].plot(x,sin2x)


ax[1,0].plot(x,cosx)


ax[1,1].plot(x,cos2x)
```
